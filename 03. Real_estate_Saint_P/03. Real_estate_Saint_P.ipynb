## Описание проекта

В нашем распоряжении данные из архива объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. 

**Задача** — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Описание данных

* `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
* `balcony` — число балконов
* `ceiling_height` — высота потолков (м)
* `cityCenters_nearest` — расстояние до центра города (м)
* `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
* `first_day_exposition` — дата публикации
* `floor` — этаж
* `floors_total` — всего этажей в доме
* `is_apartment` — апартаменты (булев тип)
* `kitchen_area` — площадь кухни в квадратных метрах (м²)
* `last_price` — цена на момент снятия с публикации
* `living_area` — жилая площадь в квадратных метрах (м²)
* `locality_name` — название населённого пункта
* `open_plan` — свободная планировка (булев тип)
* `parks_around3000` — число парков в радиусе 3 км
* `parks_nearest` — расстояние до ближайшего парка (м)
* `ponds_around3000` — число водоёмов в радиусе 3 км
* `ponds_nearest` — расстояние до ближайшего водоёма (м)
* `rooms` — число комнат
* `studio` — квартира-студия (булев тип)
* `total_area` — площадь квартиры в квадратных метрах (м²)
* `total_images` — число фотографий квартиры в объявлении

```python
# функция, которая заменяет пропущенные значения разницей между общей площадью и жилой
def kitchen_area (row):
    living_area = row['living_area']
    total_area = row['total_area']
    kitchen_area = row['kitchen_area']
    if kitchen_area is None:
        return (total_area - living_area)
    return kitchen_area
data['kitchen_area'] = data.apply(kitchen_area, axis=1)
```

Минимальная зона кухонной зоны в кухне-столовой должна быть не менее 6 кв.м. Следовательно, все значения, которые будут меньше этого значения необходимо заменить на медианные.

Большой процент пропущенных значений, присутствуют как аномально высокие значения - потолки выше 20 метров, так и аномально низкие - 1 метр.

**СП 54.13330.2016 ЗДАНИЯ ЖИЛЫЕ МНОГОКВАРТИРНЫЕ**

5.8 Высота (от пола до потолка) жилых комнат и кухни (кухни-столовой) в климатических подрайонах IA, IБ, IГ, IД и IVА, определяемых по СП 131.13330, должна быть не менее 2,7 м, а в других климатических подрайонах - не менее 2,5 м.
Высота внутриквартирных коридоров, холлов, передних, антресолей и под ними определяется условиями безопасности передвижения людей и должна составлять не менее 2,1 м.




```python
# поделим аномальные значения на 10
index_list = data.query('ceiling_height > 21 and ceiling_height < 36').index 
data.loc[index_list, 'ceiling_height'] = data.loc[index_list, 'ceiling_height'] / 10
```

```python
# заменим значения высоты потолков на общую медиану для отсутствующих значений этажности дома
for floor in ceiling_medians.index:
    data.loc[(data['total_floors'].isna()) | 
             ((data['total_floors'] == floor) & (ceiling_medians.loc[floor, 'count'] == 0)), 
             'ceiling_height'] = data['ceiling_height'].describe()[5]
```

```python
# медианы из сгруппированной таблицы вставляем на места пропущенных значений высоты потолков для дома соответствующей этажности
for floor in ceiling_medians.index:
    data.loc[(data['ceiling_height'].isna()) & 
             (data['total_floors'] == floor), 
             'ceiling_height'] = ceiling_medians.loc[floor, 'median']
```

```python
# заменим на минимальное значение по СП - 2.7 м
data.loc[(data.ceiling_height <= 2.1), 'ceiling_height'] = 2.70
data.loc[(data.ceiling_height >= 3.6), 'ceiling_height'] = 2.70
```
